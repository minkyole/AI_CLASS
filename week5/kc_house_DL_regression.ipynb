```python
df['date'] = pd.to_datetime(df['date'], format='%Y%m%dT%H%M%S')
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
df['day'] = df['date'].dt.day
df = df.drop(columns=['id', 'date'])
print(df.dtypes)
```

```python
X = df.drop('price', axis=1)
y=df['price']
y.value_counts()
```

```python
print(X.dtypes)
```

```python
X = X.values # 딥러닝 입력을 위해 numpy로 변환
```

```python
# 6. 데이터 스케일링 (특히 딥러닝 모델은 스케일된 데이터에서 잘 동작함)
scaler = StandardScaler()
X = scaler.fit_transform(X)
```

```python
X_train_val, X_test, y_train_val, y_test = train_test_split(X, y, test_size=0.15, random_state=42)
X_train, X_val, y_train, y_val = train_test_split(X_train_val, y_train_val, test_size=0.15/0.85, random_state=42)
```

```python
# 컬럼갯수 확인
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
print(X_val.shape, y_val.shape)
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam
import tensorflow as tf
from tensorflow import keras


model = Sequential()

model.add(Dense(20,input_shape=(X_test.shape[1],),activation='relu')) # 입력층 최소 입력이랑 갯수 똑같게
model.add(Dense(30,activation='relu')) # 히든층
model.add(Dense(20,activation='relu')) # 히든층
model.add(Dense(1)) # 출력층  출력이랑 갯수 똑같게
#히든층은 마름모꼴 구조로 만드는걸 교수님은 추천 혹은 피라미드 구조
# 회귀 문제라서 loss funciton mse 및 출력층 None
model.compile(optimizer='adam',loss='mse',metrics=['mae'])
# 회귀문제라서 metrics(평가지표)를 mae(예측값과 실제값 사이 평균오차)로 사용

model.summary()
```

```python
model_history = model.fit(x=X_train, y=y_train, epochs=30, batch_size=32, validation_data=(X_val, y_val))
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
# y_test: 실제값, y_pred: 모델 예측값
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("MSE:", mse)
print("MAE:", mae)
print("R² Score:", r2)
```
